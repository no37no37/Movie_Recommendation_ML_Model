```python
import pandas as pd
```

```python
data = pd.read_csv("/content/netflix_titles.csv")
```

```python
data.head(4)
```

```python
# now lets understand what are the input parameters and output parameters.

# director, cast, listed_in, description - these four columns are the input columns.
# I'll add this to a new column called "tags"

data['tags'] = data['director'] + '' + data['cast'] + '' + data['listed_in'] + '' + data['description']
```

```python
data.head(10)
```

```python
# now we have 13 columns
# lets drop columns such as director, cast, country, date_added, release_year

data = data[['show_id','type','title','tags']]
```

```python
data
```

```python
# there are null values in the 'tags'

data.dropna(inplace=True)
```

```python
data
```

```python
# the index is showing 5396 rows, but we have only 3281 rows.
# so, we need to reset index

data.reset_index(inplace=True, drop=True)
```

```python
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# CountVectorizer converts alphabetical values to numerical value, which will be later feeded into the algorithm.
```

```python
cv = CountVectorizer(max_features=3281,stop_words='english')

# So, this will create the shaoe of 3281 rows. For each and every row it'll create a vector.
# stop_words ='english' //so repeated words are ignored
```

```python
# once the CountVectorizer is ready, we have to call our algorithm.
# Algorithm we are using here is "cosine_similarity"

from sklearn.metrics.pairwise import cosine_similarity

# Cosine Similarity - It measures the similarity between two vectors by calculating the cosine of the angle between the two vectors.
# Eg: Hi World and Hello World, to find out the difference between these two..
# first the cosine_similarity will draw these two vectors in XYZ dimensional graph..
# for example Hi World will be drawn between X and Y axis. And Hello World has drawn between Z axis and Y axis..
# To find out how similar these sentences, we have to find out theta.
# Suppose if you have 100 records it will draw the 100 records in the XYZ dimensional. And then if i have to find out the similarity between first record vs 50th record, it'll find out the angle of difference between those two.
```

```python
# we have to convert our input dataset to vector.

vector = cv.fit_transform(data['tags'].values.astype('U')).toarray()

# we are preparing this 'tags'. So that is the only input column.
# based upon this 'tags', we have to identify the title.
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
# lets find out the similarity for a particular movie.
# lets take an example of index

data[data['title']=='Amrapali'].index[0]
```

```python
similarity[3278]
```

```python
# The above is an array which contains 3278 entries


# let me convert it into list and enumerate.
# then sort out in the reverse basis..
# because we need the data which is nearest to that or which is similar to whatever our input dataser
similar_movies = sorted(list(enumerate(similarity[3278])),reverse=True, key=lambda vector:vector[1])

# we got (3278, 1.0)
# which means this particular movie is 100% similar to the same movie
```

```python
for i in similar_movies[0:6]:
  print(data.iloc[i[0]].title)

# here I took top 6, For Amrapali these 5 are the similar ones, based upon the data of the 'tags'.
```

```python
# I want to find the percentage of the nearest similarity in the recommendation.
for i in similar_movies[1:6]:
  print(data.iloc[i[0]].title)
  print(i[1])


# it shows Amrapali as 100%, Wave of Cinema as 30% etc.
```

```python
# lets build a function with all this logic

def recommend_movie(movie_name):
  movie_index = data[data['title']==movie_name].index[0]
  similar_movies = sorted(list(enumerate(similarity[movie_index])),reverse=True, key=lambda vector:vector[1])
  for i in similar_movies[1:6]:
    print(data.iloc[i[0]].title +' with similarity of '+ str(round(i[1]*100,2)))
```

```python
recommend_movie('Ganglands')
```

```python
# now it is giving the top 5 recommendation.
```
